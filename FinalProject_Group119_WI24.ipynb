We hypothesize that higher education correlates with greater life satisfaction due to improved job prospects, social networks, and financial stability. However, there may be a bell curve effect: highly specialized professionals may sacrifice socializing or work-life balance, while those with much lower education levels may face challenges in job access and financial stability, leading to lower life satisfaction levels.

The General Social Survey (GSS) is an ongoing sociological survey obtained through face-to-face or telephone interviews, retrieving data on social trends and attitudes in the United States. The GSS was first conducted in 1972, and until 1994, conducted almost annually. Now, it is conducted in even numbered years. 


The target population of this survey are adults (18+) living in any kind of household, and languages offered are either English or Spanish (offered 2006 and on). However, residents of institutions and group quarters are not included in the survey. Addresses are specifically chosen to represent a cross-section sample of households __*__. Households are then randomly selected across the country to ensure scientific validity, since each household has an equal chance of being selected. In re-interviews, individuals who have passed away, moved out of the country, or no longer lived in a household are outside of the target population. Therefore, these individuals do not participate again. 


Among the more than 5000 variables within the database, we have selected `educ`, which is a numerical variable representing the highest year of school completed by the respondent, and four satisfaction-related factors: `satjob`, `satfin`, `satsoc`, and `satlife`. These ordinal variables represent the respondent's satisfaction with their work, financial situation, social activities and relationships, and life as a whole, respectively. 


The questions posed in the survey allowed the respondent to choose an answer on the basis of a scale; for example, a question for `satjob` would ask “How satisfied are you with the work you do? Satisfied, moderately satisfied, a little dissatisfied, or very dissatisfied?” Each of these variables relevant to satisfaction-related factors included either the same or similar scale of 4-6 different options. The table below shows the different options in the scale of each variable: 


| Variable  | Scale Options |
|---|---|
| `satjob`  |  very satisfied, moderately satisfied, a little dissatisfied, or very dissatisfied |
| `satfin`  |  pretty well satisfied, more or less satisfied, not satisfied at all |
| `satsoc` |  poor, fair, good, very good, excellent |
| `satlife` | completely dissatisfied, mostly dissatisfied, somewhat dissatisfied, Neither satisfied nor dissatisfied, somewhat satisfied, mostly dissatisfied, completely satisfied |


The data was extracted from the GSS Data Explorer tool for the year 2018 because this was the most recent year which had the most filled-out survey responses (rather than empty, non-applicable answers) for each satisfaction factor variable. For the year 2018, there were a total of 2348 responses. Other years either were much more distant, raising concerns about a confounder, cultural divergence across time periods, or did not have consistent, completed survey responses across life satisfaction variables.__*__


In terms of extraction process, relevant variables were searched for (`educ`, `satjob`, `satfin`, `satsoc`, and `satlife`) and the “Add to Extract” option was selected. Then, on the Extract page on the user’s profile, a new extract was named and the same variables were added and filtered to 2018, and formatted as an Excel workbook with labeled metadata to allow reproducibility.

```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import patsy
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data
gss = pd.read_csv("GSS.csv")
gss.head()
```

```python
# Drop year and ID columns
gss = gss.drop(columns=['year', 'id_'])
```

```python
# Renaming the columns of the dataframe to more understandable names
gss.columns = ['education_level', 'job_satisfaction', 'financial_satisfaction', 
               'social_satisfaction', 'life_satisfaction']
```

The data is relatively clean as we have extracted our data with the help of GSS Data Explorer. However, we do have some 'inapplicable', 'no answer', and 'do not know' entries in the dataset that need to be processed.

```python
# Function to change 'inapplicable' / 'no answer' / 'do not know' answers to NaN

def standardize_invalid_answer(string):
    if '.i:' in string:     # if answer is inapplicable, change to NaN
        output = np.nan
    elif '.n:' in string:   # if answer was not given, change to NaN
        output = np.nan
    elif '.d:' in string:   # if answer was unknown, change to NaN
        output = np.nan
    else:
        output = string
    return output
```

```python
# Change 'inapplicable' / 'no answer' / 'do not know' answers to NaN
gss['education_level'] = gss['education_level'].apply(standardize_invalid_answer)
gss['job_satisfaction'] = gss['job_satisfaction'].apply(standardize_invalid_answer)
gss['financial_satisfaction'] = gss['financial_satisfaction'].apply(standardize_invalid_answer)
gss['social_satisfaction'] = gss['social_satisfaction'].apply(standardize_invalid_answer)
gss['life_satisfaction'] = gss['life_satisfaction'].apply(standardize_invalid_answer)
```

```python
# Check for invalid answers
gss['job_satisfaction'].value_counts()
```

Next, we will remove rows with NaN values from our dataset to ensure accuracy in our analysis. NaN values indicate missing data, which could distort our results and complicate the analysis process. By removing these rows, we’re ensuring that our analysis is based on complete and reliable data.

```python
# Drop the rows with missing data and reset the index
gss = gss.dropna().reset_index(drop=True)
```

## Analysis: Linear Regression of Each Variable

After the stacked bar plots, we decided to do linear regression on each variable to see if there is any correlation between education level and each of the satisfaction levels.

We will first convert the categorical labels representing satisfaction levels in different aspects of life into numerical values for the columns `job_satisfaction`, `financial_satisfaction`, `social_satisfaction`, and `life_satisfaction`. Beginning at a level of 1 for the lowest satisfaction level, the higher the satisfaction, the higher the numerical value. For example, in the `job_satisfaction` column, the label 'Very dissatisfied' is mapped to 1, 'A little dissatisfied' to 2, and so on. We then rescale these numerical values to a common range of 0-10 using the min-max normalization technique. This step is necessary because the four satisfaction columns may not be on the same scale initially, and converting them to a common range allows for better comparison and analysis.

```python
# Convert labels to values

# Functions that maps a label to a numerical value
job_label_to_num = {
    'Very dissatisfied': 1,
    'A little dissatisfied': 2,
    'Moderately satisfied': 3,
    'Very satisfied': 4
}

fin_label_to_num = {
    'Not satisfied at all': 1,
    'More or less satisfied': 2,
    'Pretty well satisfied': 3
}

soc_label_to_num = {
    'Poor': 1,
    'Fair': 2,
    'Good': 3,
    'Very good': 4,
    'Excellent': 5
}

lif_label_to_num = {
    'Completely dissatisfied': 1,
    'Mostly dissatisfied': 2,
    'Somewhat dissatisfied': 3,
    'Neither satisfied nor dissatisfied': 4,
    'Somewhat satisfied': 5,
    'Mostly satisfied': 6,
    'Completely satisfied': 7
}

# Implement the functions on the DataFrame
gss['job_satisfaction'] = gss['job_satisfaction'].replace(job_label_to_num).astype(float)
gss['financial_satisfaction'] = gss['financial_satisfaction'].replace(fin_label_to_num).astype(float)
gss['social_satisfaction'] = gss['social_satisfaction'].replace(soc_label_to_num).astype(float)
gss['life_satisfaction'] = gss['life_satisfaction'].replace(lif_label_to_num).astype(float)

# Rescale the numerical values to the range 0 to 10
job_min = gss['job_satisfaction'].min()
job_max = gss['job_satisfaction'].max()
gss['job_satisfaction'] = (gss['job_satisfaction'] - job_min) / (job_max - job_min) * 10

fin_min = gss['financial_satisfaction'].min()
fin_max = gss['financial_satisfaction'].max()
gss['financial_satisfaction'] = (gss['financial_satisfaction'] - fin_min) / (fin_max - fin_min) * 10

soc_min = gss['social_satisfaction'].min()
soc_max = gss['social_satisfaction'].max()
gss['social_satisfaction'] = (gss['social_satisfaction'] - soc_min) / (soc_max - soc_min) * 10

lif_min = gss['life_satisfaction'].min()
lif_max = gss['life_satisfaction'].max()
gss['life_satisfaction'] = (gss['life_satisfaction'] - lif_min) / (lif_max - lif_min) * 10

#Display the new DataFrame
gss.head()
```

For each of the satisfaction variables (`job_satisfaction`, `financial_satisfaction`, `social_satisfaction`, and `life_satisfaction`), we will draw an exploratory plot with the education level as the independent variable on the x-axis and the average satisfaction for each given education level as the dependent variable on the y-axis. In order to further measure the relationship, we will use linear regression models for each of the satisfaction variables with education level as the independent variable. We assume that education level and each of the satisfaction variables have a linear relationship. As we are only working with education level as the independent variable and satisfaction variables as the dependent variables, we don't have any multicollinearity concerns. Each point in the scatter plot will represent the average level of satisfaction for a specific education level, so they are all independent observations, and there is no autocorrelation. Finally, the data have a similar variance across the entire distribution, which implies that the data is homoscedastic. With the assumptions above, we can safely carry out linear regression analysis.

### Job Satisfaction Analysis

```python
# Plot the scatterplot for average satisfaction for each given education level
avg_job_sat = gss.groupby('education_level')['job_satisfaction'].mean()
df_job = avg_job_sat.reset_index()
sns.scatterplot(x='education_level', y='job_satisfaction', data=df_job, color='red')  
plt.xlabel('Education Level')
plt.ylabel('Average Job Satisfaction')
plt.title('Average Job Satisfaction by Education Level')
plt.show()
```

Looking at the scatterplot above, we can see that there is a positive correlation. Now let's look at the regression results to look at the significance.

```python
# Regression results
outcome, predictors = patsy.dmatrices('job_satisfaction ~ education_level', df_job) 
job_model = sm.OLS(outcome, predictors)
job_results = job_model.fit()
print(job_results.summary())
```

As shown above, we converted our labels to values. In doing so, a low value average of job satisfaction would represent that the average level of satisfaction within that particular education level is more unsatisfactory. On the other hand, a high value average of job satisfaction would represent that the average level of satisfaction within that particular education level is more satisfactory. Looking at this variable alone, we find that the effect size to be 0.1090.

```python
# Plot the regression line for the scatterplot
sns.scatterplot(x='education_level', y='job_satisfaction', alpha=0.3, data=df_job, color='red')

xs_job = np.arange(df_job['education_level'].min(), df_job['education_level'].max())
ys_job = job_results.params[1] * xs_job + job_results.params[0] 
plt.plot(xs_job, ys_job, '--k', label='Model')

plt.xlabel('Education Level')
plt.ylabel('Average Job Satisfaction')
plt.legend()
plt.show()
```

Considering this scatterplot, we confer that there is a positive relationship between average job satisfaction and education. Thus, as education levels increase, the level of job satisfaction also increases. Moreover, we find this variable to be statistically significant as the p-value is 0.020 and moderately correlated with education level due to the r-value of 0.403. In doing so, we confer that there is a significant effect between education levels and job satisfaction.

### Financial Satisfaction Analysis

```python
# Plot the scatterplot for average satisfaction for each given education level
avg_fin_sat = gss.groupby('education_level')['financial_satisfaction'].mean()
df_fin = avg_fin_sat.reset_index()
sns.scatterplot(x='education_level', y='financial_satisfaction', data=df_fin, color='blue')  
plt.xlabel('Education Level')
plt.ylabel('Average Financial Satisfaction')
plt.title('Average Financial Satisfaction by Education Level')
plt.show()
```

This scatterplot seems to also have a positive correlation, however, we should still check the significance level and the line of best fit.

```python
# Regression results
outcome, predictors = patsy.dmatrices('financial_satisfaction ~ education_level', df_fin) 
fin_model = sm.OLS(outcome, predictors)
fin_results = fin_model.fit()
print(fin_results.summary())
```

Looking at the variable `financial_satisfaction` alone, we find the effect size to be 0.2089, which is higher in comparison to the job satisfaction variable. Moreover, we find this variable to also be statistically significant with a p-value of 0.000, and has a higher correlation with education level in comparison to the job satisfaction variable as the r-value is 0.729. Conclusively, there is an effect between education levels and financial satisfaction.

```python
# Plot the regression line for the scatterplot
sns.scatterplot(x='education_level', y='financial_satisfaction', alpha=0.3, data=df_fin, color='blue')

xs_fin = np.arange(df_fin['education_level'].min(), df_fin['education_level'].max())
ys_fin = fin_results.params[1] * xs_fin + fin_results.params[0] 
plt.plot(xs_fin, ys_fin, '--k', label='Model')

plt.xlabel('Education Level')
plt.ylabel('Average Financial Satisfaction')
plt.legend()
plt.show()
```

Looking at the scatterplot for the variable `financial_satisfaction`, a low average value of financial satisfaction assumes that the respondents within the particular education level are less satisfied with their financial circumstances; while a high average value assumes that the respondents within the particular education level are more satisfied with their financial circumstance. As the level of education increases, the level of satisfaction within one’s financial situation increases as well.

# Ethics & Privacy

As we are utilizing the General Social Survey (GSS), this database is open to the public to access. The survey respondents’ personal identifiable information (PII) is protected (i.e. name, address, etc) and GSS has a protocol that ensures the privacy of participants and their information (and if PII is necessary, acquiring that personal data would require a thorough application process by researchers). Overall, the dataset we’ve proposed and the variables included in our analysis ensure that the information of these human subjects is kept confidential.

However, in utilizing this dataset, we are met with potential biases that only include a particular target population: adults (18+ individuals) who live in a household. (__Note:__ living in a household doesn’t necessarily define an individual who owns a home. Moreover, this survey excludes populations who reside in institutions such as college dorms and group quarters). This survey is limited to people who speak a certain language (English or Spanish), and information on race only allows respondents to choose “White”, “Black”, or “Other”. In doing so, the analysis of this data is limited to this particular subset of people and potentially excludes identities of other cultural groups, ethnicities, or circumstances. Moreover, because the dataset includes self-reported data, we risk the accuracy of information as we are unaware of the environment in which the respondent took the survey and how it affected the honesty of their responses.

As we are limited to a particular demographic, we acknowledge that our analysis is not fully representative of other cultural/ethnic identities who could have different life circumstances. However, by having this narrow population, we are able to reduce our analysis and provide insight into this particular demographic. In handling these issues mentioned above, it’s crucial to be able to communicate the limitations included within our data and point out what we are specifically analyzing.

# Discussion and Conclusion

Our goal for this data project was to investigate the relationship between the level of education and overall life satisfaction in the US. We explored if specific areas of satisfaction differed significantly based on varying education levels, ultimately proposing our hypothesis: higher education does correlate with greater life satisfaction, but a bell curve effect would be shown with much lower satisfaction for those with the lowest and highest satisfaction levels. 
Seeking to prove if our hypothesis was correct, we extracted data from the General Social Survey’s 2018 responses under life satisfaction-related factors, such as satisfaction with work, financial situation, social activities, and life overall, as well as respondents’ education levels. __After data analysis, our finding was that there is support for a positive relationship between education level and overall life satisfaction in the US, rejecting the second part of our hypothesis proposing a bell curve effect.__

Applying linear regression analysis, we found that each test for all four life satisfaction variables had a p-value < 0.02, indicating a rejection of the null hypothesis; there is a significant statistical correlation between life satisfaction and education level. Also, each satisfaction variable had a positive correlation with education levels, but financial satisfaction had the greatest impact on overall satisfaction, with the highest effect size and R-squared value being 0.2809 and 0.729, respectively, within our sample population of 2348 respondents.




_Limitations and Impact on Society_

However, this analysis raises some limitation concerns, with the first being that 0.2809 is a small effect size. Regardless, though our findings may not be considered substantial, there is still a noticeable effect of education levels affecting life satisfaction levels. Another consideration is that both job satisfaction and overall life satisfaction analyses began with relatively high levels with a y-intercept of 6.177 and 6.153, irrespective of education levels, indicating pre-existing baselines of satisfaction under these domains. Additionally, most of our R-squared values ranged from 0.403 to 0.494. The R-squared value of 0.403 for average job satisfaction indicates that only 40.3% of the variance in job satisfaction can be explained by education level. This indicates only a moderate positive relationship, meaning that 59.7% of the variability remains undiscovered and possibly suggests other cofounders at play.

This positive correlation in our findings could be due to the value education plays in American society: a higher level of education could result in high-paying job opportunities, thus leading to a comfortable lifestyle. Taking into account that certain variables, such as job satisfaction and life satisfaction, have high y-intercept values, this would indicate that individuals with lower levels of education are at least still moderately satisfied with their occupation and life. Moreover, we found a potential limitation with the validity of these responses as we acquired self-reported data. In doing so, the survey responses are more subjective and introspective rather than fact-based. We acknowledge that different individuals have varying scales of life satisfaction, and miscellaneous confounders such as cultural, financial, and geographical background also play a role in affecting limitations of education level and life satisfaction variables. If we were to improve upon our study in the future, we would be interested to see how the data would look over a certain time period– for example, comparing the data from the last ten years. We would also be interested in analyzing the differences between cultures and ethnicities and how they would differ between these confounding variables.

Ultimately, drawing back to our background information (Teachers College survey indicating widespread belief in the benefits of higher education, a Pew Research study highlighting the perceived usefulness of college education among graduates, Gallup data unveiling significant proportions of Americans expressing regret over their educational choices, and study in Australia finding a positive correlation between university completion and factors influencing life satisfaction) we discovered that our analyses generally supports these prior works. As individuals move forward to higher levels of education, we find that these satisfaction levels rise because these people are continuously discovering their interests and passions over time. The Gallup data highlights that some individuals who held a Bachelor’s degree regretted their decision to their field of study as opposed to those who possess a postsecondary education. Thus, in relation to our study, we see that it embodies this idea of continuously pursuing an education for the topics and ideas that interest people the most. In doing so, we see that people who have the means to pursue this opportunity could lead a more fulfilling life. Education provides opportunities, not only for financial gain but internal satisfaction of purpose.
